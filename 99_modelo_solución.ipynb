```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from google.colab import userdata
```

```python
students_df = pd.read_csv("train.csv")
```

```python
id_estudiantes = students_df["ID"].copy()
```

```python
students_df = students_df.drop(['ID'], axis=1)
students_df = students_df.fillna('Miss')
```

```python
def to_onehot(x):
    values = np.unique(x)
    r = np.r_[[np.argwhere(i==values)[0][0] for i in x]]
    return np.eye(len(values))[r].astype(int)

def replace_column_with_onehot(d, col):
    assert sum(d[col].isna())==0, "column must have no NaN values"
    values = np.unique(d[col])
    k = to_onehot(d[col].values)
    r = d.join(pd.DataFrame(k, columns=["%s_%s"%(col, values[i]) for i in range(k.shape[1])], index=d.index), how='outer')
    del(r[col])
    return r

columns_to_onehot =[
    'F_TIENEINTERNET',
    'E_PAGOMATRICULAPROPIO',
]
onehoted_df = students_df.copy()
for col in columns_to_onehot:
  onehoted_df = replace_column_with_onehot(onehoted_df, col)
onehoted_df
```

```python
def replace_category_for_numbers(d, col, vals_to_change):
  r = d.copy()
  r[col] = r[col].replace(vals_to_change.keys(), vals_to_change.values())
  return r
```

```python
onehoted_df = onehoted_df.drop(['F_TIENEINTERNET_Miss'], axis=1)
print (onehoted_df.shape)
onehoted_df
```

```python
final_student_df = replace_column_with_onehot(onehoted_df, 'E_PRGM_DEPARTAMENTO')
y = final_student_df["RENDIMIENTO_GLOBAL"].values
X = final_student_df.drop(['RENDIMIENTO_GLOBAL'], axis=1)
X.E_PRGM_ACADEMICO = X.E_PRGM_ACADEMICO.astype("category")
X.E_VALORMATRICULAUNIVERSIDAD = X.E_VALORMATRICULAUNIVERSIDAD.astype("category")
X.PERIODO_ACADEMICO = X.PERIODO_ACADEMICO.astype("category")
del final_student_df
print (X.shape, y.shape)
```

```python
id_estudiantes = id_estudiantes.loc[X.index]
```

```python
def fit_and_score(estimator, X_train, X_test, y_train, y_test):
    """Fit the estimator on the train set and score it on both sets"""
    estimator.fit(X_train, y_train, eval_set=[(X_test, y_test)])

    train_score = estimator.score(X_train, y_train)
    test_score = estimator.score(X_test, y_test)

    return estimator, train_score, test_score
```

```python
# Convertir columnas tipo object a category
for col in X.columns:
    if X[col].dtype == 'object':
        X[col] = X[col].astype('category')
```

```python
# Ensure y is a pandas Series for easier manipulation and value counts.
# The original y might be a numpy array or a Series. Converting it explicitly.
y_original_series = pd.Series(y)

# Convert 'Miss' to -1 and then to integer.
y_processed = y_original_series.replace('Miss', -1).astype(int)

# Identify classes with only one sample
class_counts = y_processed.value_counts()
problematic_classes = class_counts[class_counts == 1].index.tolist()

if problematic_classes:
    print(f"Warning: The following classes have only one member and cannot be used for stratified sampling: {problematic_classes}")
    print("These samples will be removed from the dataset before splitting.")

    # Get the indices of samples to keep (i.e., those not belonging to problematic classes)
    indices_to_keep = ~y_processed.isin(problematic_classes)

    # Filter X and y_processed
    X_filtered = X.loc[indices_to_keep]
    y_filtered = y_processed.loc[indices_to_keep].values # Convert Series to numpy array for train_test_split

    print(f"Original X shape: {X.shape}, Filtered X shape: {X_filtered.shape}")
    print(f"Original y shape: {y_processed.shape}, Filtered y shape: {y_filtered.shape}")

    # Use the filtered data for train_test_split
    Xtr, Xts, ytr, yts = train_test_split(
        X_filtered, y_filtered,
        test_size=0.2,
        random_state=42,
        stratify=y_filtered
    )
else:
    print("No problematic classes with only one member found after conversion.")
    # If no problematic classes, just use the converted y for splitting
    Xtr, Xts, ytr, yts = train_test_split(
        X, y_processed.values, # Use the converted y as numpy array
        test_size=0.2,
        random_state=42,
        stratify=y_processed.values
    )

print("Datos divididos:")
print(f"- Entrenamiento: X={Xtr.shape}, y={ytr.shape}")
print(f"- Prueba: X={Xts.shape}, y={yts.shape}")
```

```python
import numpy as np

estimator_rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    random_state=42,
    n_jobs=-1
)

print("Iniciando entrenamiento...")

# Ensure Xtr and Xts are fully numerical before fitting

# 1. Handle INDICADOR columns: Replace 'Miss' with NaN and convert to float
indicator_cols = ['INDICADOR_1', 'INDICADOR_2', 'INDICADOR_3', 'INDICADOR_4']
for col in indicator_cols:
    if col in Xtr.columns:
        # Use .copy() to avoid SettingWithCopyWarning
        Xtr[col] = pd.to_numeric(Xtr[col].replace('Miss', np.nan), errors='coerce')
        Xts[col] = pd.to_numeric(Xts[col].replace('Miss', np.nan), errors='coerce')

# 2. Convert all remaining categorical columns to one-hot encoding
# pd.get_dummies will handle 'category' and 'object' dtypes
Xtr_processed = pd.get_dummies(Xtr)
Xts_processed = pd.get_dummies(Xts)

# 3. Align columns - crucial for consistent feature sets between train and test
# This step ensures both train and test sets have the same columns in the same order
# and handles categories present in one but not the other.

# Get all unique columns from both processed DataFrames
all_cols = Xtr_processed.columns.union(Xts_processed.columns)

# Reindex both DataFrames to have all_cols, filling new columns with 0
Xtr_processed = Xtr_processed.reindex(columns=all_cols, fill_value=0)
Xts_processed = Xts_processed.reindex(columns=all_cols, fill_value=0)

# 4. Fill any NaNs that might have been introduced (e.g., from INDICADOR columns)
# For Random Forest, mean imputation is a common strategy, or keep NaNs if the model supports it.
# Let's fill with the mean for simplicity, using train's mean for both train and test.
for col in Xtr_processed.columns:
    if Xtr_processed[col].isnull().any():
        mean_val = Xtr_processed[col].mean()
        Xtr_processed[col] = Xtr_processed[col].fillna(mean_val)
        Xts_processed[col] = Xts_processed[col].fillna(mean_val)


# Now fit the estimator with the fully numerical and aligned data
estimator_rf.fit(Xtr_processed, ytr)

print("Entrenamiento finalizado.")
```

```python
y_pred_rf = estimator_rf.predict(Xts_processed)

# Scores de precisión (Accuracy)
train_score_rf = estimator_rf.score(Xtr_processed, ytr)
test_score_rf = estimator_rf.score(Xts_processed, yts)

print(f"Accuracy de entrenamiento (In-Sample): {train_score_rf:.4f}")
print(f"Accuracy de prueba (Out-of-Sample): {test_score_rf:.4f}")
```

```python
# Diccionario de clases
data_performance_dict = {
    0: 'bajo',
    1: 'medio-bajo',
    2: 'medio-alto',
    3: 'alto',
}

# Se convierte las predicciones numéricas a texto
y_labels = [data_performance_dict[val] for val in y_pred_rf]

# Se recuperan los IDs reales del test
ids = id_estudiantes.loc[Xts.index].values

# Se crea el DataFrame de salida
submission = pd.DataFrame({
    "ID": ids,
    "RENDIMIENTO_GLOBAL": y_labels
})

# Exportamos el archivo
submission.to_csv("out.csv", index=False)
```

```python
print('Loading test data...')
test_df = pd.read_csv('test.csv')

# Store original IDs for submission
test_ids = test_df['ID'].copy()

# Drop ID column from test set
test_df = test_df.drop(['ID'], axis=1)

# Fill NaN values with 'Miss' (same as training data)
test_df = test_df.fillna('Miss')
print('Test data loaded and initial processing complete.')
```

```python
print('Applying one-hot encoding for specific columns on test data...')
onehoted_test_df = test_df.copy()
for col in columns_to_onehot:
  onehoted_test_df = replace_column_with_onehot(onehoted_test_df, col)

print('Applying category-to-number mapping on test data...')
# Ensure we don't try to convert 'RENDIMIENTO_GLOBAL' for the test set
test_data_to_convert = {k: v for k, v in data_to_convert.items() if k != 'RENDIMIENTO_GLOBAL'}
for col, transform_dict in test_data_to_convert.items():
    onehoted_test_df = replace_category_for_numbers(
        onehoted_test_df,
        col,
        transform_dict)

# Drop F_TIENEINTERNET_Miss (same as training data)
onehoted_test_df = onehoted_test_df.drop(['F_TIENEINTERNET_Miss'], axis=1)

# One-hot encode E_PRGM_DEPARTAMENTO (same as training data)
final_test_df = replace_column_with_onehot(onehoted_test_df, 'E_PRGM_DEPARTAMENTO')

print('Test data preprocessing (one-hot and category-to-number) complete.')
```

```python
print('Converting object columns to category and handling INDICADOR columns in test data...')

# Convert object columns to category (same as training data)
for col in final_test_df.columns:
    if final_test_df[col].dtype == 'object':
        final_test_df[col] = final_test_df[col].astype('category')

# Handle INDICADOR columns: Replace 'Miss' with NaN and convert to float (same as training data)
indicator_cols = ['INDICADOR_1', 'INDICADOR_2', 'INDICADOR_3', 'INDICADOR_4']
for col in indicator_cols:
    if col in final_test_df.columns:
        final_test_df[col] = pd.to_numeric(final_test_df[col].replace('Miss', np.nan), errors='coerce')

print('Final preprocessing steps for test data complete.')
```

```python
print('Applying one-hot encoding and aligning columns for test data...')

# Apply get_dummies to the test data
X_test_processed = pd.get_dummies(final_test_df)

# Get all columns from the processed training data (Xtr_processed)
all_training_cols = Xtr_processed.columns

# Reindex the test data to align with training columns, filling new columns with 0
X_test_processed = X_test_processed.reindex(columns=all_training_cols, fill_value=0)

# Impute NaNs using means from the training data
# Calculate means from Xtr_processed once
imputation_means = Xtr_processed.mean(numeric_only=True)

for col in X_test_processed.columns:
    if X_test_processed[col].isnull().any():
        if col in imputation_means:
            X_test_processed[col] = X_test_processed[col].fillna(imputation_means[col])
        else:
            # Fallback for columns not in training means (e.g., if a new column appeared unexpectedly)
            X_test_processed[col] = X_test_processed[col].fillna(0) # Or another appropriate default

print('Column alignment and imputation for test data complete.')
```

```python
print('Generating predictions on the preprocessed test data...')

y_pred_test = estimator_rf.predict(X_test_processed)

# Convert numerical predictions to text labels
y_labels_test = [data_performance_dict[val] for val in y_pred_test]

# Create the submission DataFrame
submission_df = pd.DataFrame({
    "ID": test_ids,
    "RENDIMIENTO_GLOBAL": y_labels_test
})

# Export the submission file
submission_df.to_csv("out.csv", index=False)

print(f"Submission file 'out.csv' created with {len(submission_df)} rows.")
print('You can now download out.csv and submit it to Kaggle.')
```
